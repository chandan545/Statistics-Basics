1. Explain the different types of data (qualitative and quantitative) and provide examples of each. Discuss
nominal, ordinal, interval, and ratio scales.
   - Data types can be broadly categorized into qualitative and quantitative data. These types are used to classify, analyze, and interpret information.
     - Qualitative Data
      - Qualitative data describes characteristics, attributes, or categories and is non-numeric in nature. It is often used to categorize or label observations.
      - Examples:
        - Nominal Scale: Data that categorizes without any order.
          - Examples: Gender (male, female), colors (red, blue, green), types of vehicles (car, bike, bus).
        - Ordinal Scale: Data that categorizes with a meaningful order or rank but without consistent differences between categories.
          - Examples: Satisfaction levels (satisfied, neutral, dissatisfied), education levels (high school, bachelorâ€™s, masterâ€™s).
     - Quantitative Data
      - Quantitative data represents measurable quantities and is numerical in nature. It can be analyzed mathematically.
      - Examples:
        - Interval Scale: Numerical data where the intervals between values are equal, but there is no true zero point.
          - Examples: Temperature in Celsius or Fahrenheit, IQ scores.
        - Ratio Scale: Numerical data with equal intervals and a true zero point, allowing for meaningful ratios.
          - Examples: Weight, height, age, income.

2. What are the measures of central tendency, and when should you use each? Discuss the mean, median,
and mode with examples and situations where each is appropriate.
  - Measures of central tendency summarize a dataset by identifying a single value that represents the center of its distribution. The main measures are mean, median, and mode, each suitable for specific scenarios.
  - Mean (Average)
    - The sum of all data points divided by the number of data points.
    - When to Use:
      - Data is quantitative.
      - The dataset has no extreme outliers (outliers can skew the mean).
    - Examples of Use:
      - Average test scores of a class.
      - Average income of a population.
  - Median
    - Definition: The middle value when data points are ordered. If the dataset has an even number of values, the median is the average of the two middle values.
    - When to Use:
      - Data is quantitative or ordinal.
      - The dataset has outliers or is skewed (median is resistant to outliers).
    - Examples of Use:
      - Median house prices in a city.
      - Median age of a group of people.
  - Mode
    - Definition: The most frequently occurring value(s) in the dataset. A dataset may have one mode (unimodal), more than one mode (bimodal or multimodal), or no mode.
    - When to Use:
      - Data is categorical, qualitative, or quantitative.
      - You need to identify the most common category or value.
    - Examples of Use:
      - Most popular ice cream flavor in a survey.
      - Most frequent shoe size sold in a store.
  - Key Considerations:
    - Use the mean for balanced datasets.
    - Use the median when outliers are present.
    - Use the mode for identifying the most common value, especially in categorical data.

3. Explain the concept of dispersion. How do variance and standard deviation measure the spread of data?
   - Dispersion refers to how spread out the data values are around the central value (mean or median). It shows the variability or consistency in a dataset.
   - Variance:
     - Definition: Measures the average squared distance of data points from the mean.
     - Key Point: Larger variance means greater spread.
   - Standard Deviation:
     - Definition: The square root of the variance. It shows the average distance of data points from the mean.
     - Standard deviation gives a more intuitive measure of spread in the same units as the data.
     - Key Point: Easier to interpret since it's in the same units as the data.

4. What is a box plot, and what can it tell you about the distribution of data?
   - A box plot (or whisker plot) is a graphical representation of a dataset's distribution. It displays the following key features:
     - Median: The middle value (line inside the box).
     - Quartiles:
       - Q1 (lower quartile): 25th percentile.
       - Q3 (upper quartile): 75th percentile.
       - The box spans Q1 to Q3 (interquartile range, IQR).
     - Whiskers: Extend from the box to the smallest and largest values within 1.5Ã—IQR from the quartiles.
     - Outliers: Data points outside the whiskers are shown as individual dots.
     - What It Tells You:
       - Spread: The range and variability of data.
       - Center: The median value.
       - Skewness: Symmetry or asymmetry in the data.
       - Outliers: Unusual or extreme values.
     - Example:
       - A box plot can show if data is symmetric, skewed, or has outliers, making it great for comparing distributions.

5. Discuss the role of random sampling in making inferences about populations.
   - Random sampling is a method where every individual in a population has an equal chance of being selected. It helps ensure the sample represents the population, allowing us to make accurate inferences.
   - Role:
    - Representativeness: Reflects the population's diversity.
    - Generalization: Allows findings from the sample to apply to the population.
    - Accuracy: Reduces bias and sampling error.
    - Statistical Validity: Supports reliable statistical tests and predictions.
   - Example: To estimate average height in a school, random sampling ensures students from all grades are included fairly, avoiding bias.

6. Explain the concept of skewness and its types. How does skewness affect the interpretation of data?
   - Skewness measures the asymmetry of a data distribution.  It indicates whether the data is more concentrated on one side of the mean, and helps to understand the shape of the distribution.
   - Types:
     - Positive Skew (Right Skew): Right tail is longer. Mean > Median.
       - Example: Income distribution.
     - Negative Skew (Left Skew): Left tail is longer. Mean < Median.
       - Example: Age at retirement.
     - No Skew: Symmetrical distribution. Mean = Median.
   - How Skewness Affects Data Interpretation:
     - Positive Skew: Mean is higher than the median.
     - Negative Skew: Mean is lower than the median.
     - Skewness affects which central measure (mean or median) is more appropriate to describe the data.

9. Explain the properties of the normal distribution and the empirical rule (68-95-99.7 rule).
   - Properties of the Normal Distribution:
     - Bell-shaped curve: Symmetrical and centered around the mean.
     - Mean = Median = Mode: All are equal and located at the center.
     - Defined by Mean and Standard Deviation: Shape depends on these two parameters.
     - Asymptotic: The curve approaches but never touches the x-axis.
     - Total Area = 1: Represents 100% of the data.
   - Empirical Rule (68-95-99.7 Rule):
     - For a normal distribution:
       - 68% of data falls within 1 standard deviation of the mean.
       - 95% of data falls within 2 standard deviations of the mean.
       - 99.7% of data falls within 3 standard deviations of the mean.
   - This rule helps estimate probabilities and identify outliers.

11. Explain what a random variable is and differentiate between discrete and continuous random variables.
    - A random variable is a numerical value that represents the outcome of a random event.
    - Types:
      - Discrete Random Variable:
        - Takes specific, countable values (e.g., 0, 1, 2, ...).
        - Example: Number of heads in 5 coin flips.
      - Continuous Random Variable:
        - Takes any value within a range (infinite possibilities).
        - Example: Height of students in a class.
    - Key Difference: Discrete variables are countable, while continuous variables are measurable.
